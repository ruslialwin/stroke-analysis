```python
# fitur age(umur) menunjukkan korelasi yang lebih tinggi dibandingkan dengan fitur lain, 
# sehingga nilai bmi yang kosong akan diisi menggunakan teknik imputasi berdasarkan umur

# kategorikan usia dengan binning, nilai bins ditentukan berdasarkan domain knowledge â†’ contoh di medis/WHO, 
# umur biasanya dibagi: anak-anak (<18), dewasa muda (18â€“35), dewasa menengah (36â€“50), lanjut usia (51â€“65), lansia (65+).
df["age_group"] = pd.cut(df["age"], 
                         bins=[0, 18, 35, 50, 65, 100], 
                         labels=["0-18", "19-35", "36-50", "51-65", "66+"])

# Isi NaN BMI dengan median BMI sesuai kelompok umur, median dipilih karena lebih robust terhadap outlier dibanding mean.
df["bmi"] = df.groupby("age_group")["bmi"].transform(
    lambda x: x.fillna(x.median())
)
```

```python
# Cek missing values
df.isnull().sum()
```

```python
# menghapus kolom 'id' karena tidak diperlukan dalam proses
df = df.drop(['id'], axis=1)
```

```python
# --- Fungsi untuk membuat barplot proporsi stroke ---
def plot_stroke_proportion(data, col, ax):
    # Hitung jumlah kasus stroke & total per kategori
    prop = (
        data.groupby(col)['stroke']
        .agg(['sum', 'count'])   # hitung jumlah stroke & total
        .reset_index()
    )
    prop['proporsi'] = (prop['sum'] / prop['count']) * 100  # hitung proporsi %

    # Plot barplot
    sns.barplot(x=col, y='proporsi', data=prop, hue=col, palette='viridis', ax=ax, legend=False)
    ax.set_title(f"{col} vs Stroke (Proporsi %)")
    ax.set_ylabel("Proporsi Stroke (%)")
    ax.set_xlabel(col)
    
    # Kasih label persentase di atas bar
    for container in ax.containers:
        ax.bar_label(container, fmt="%.2f%%")

# --- Daftar variabel kategorikal yang mau dicek ---
cols = ['gender', 'hypertension', 'heart_disease', 
        'ever_married', 'work_type', 'Residence_type', 
        'smoking_status']

# --- Bikin subplot ---
fig, axes = plt.subplots(2, 4, figsize=(30, 16))
axes = axes.flatten()

for i, col in enumerate(cols):
    plot_stroke_proportion(df, col, axes[i])

plt.tight_layout()
plt.show()
```

```python
# --- Ringkasan faktor berpengaruh dengan bar chart ---

factors = {
    "Hypertension": df["stroke"][df["hypertension"]==1].sum() / df[df["hypertension"]==1].shape[0],
    "Heart Disease": df["stroke"][df["heart_disease"]==1].sum() / df[df["heart_disease"]==1].shape[0],
    "Smoking (ever smoked)": df["stroke"][df["smoking_status"]=="formerly smoked"].sum() / df[df["smoking_status"]=="formerly smoked"].shape[0],
    "Smoking (smokes)": df["stroke"][df["smoking_status"]=="smokes"].sum() / df[df["smoking_status"]=="smokes"].shape[0],
    "Age > 60": df["stroke"][df["age"]>60].sum() / df[df["age"]>60].shape[0],
    "Glucose > 120": df["stroke"][df["avg_glucose_level"]>120].sum() / df[df["avg_glucose_level"]>120].shape[0],
}

# Konversi ke DataFrame
factors_df = pd.DataFrame({
    "Factor": list(factors.keys()),
    "Stroke Rate": list(factors.values())
})

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x="Stroke Rate", y="Factor", hue="Factor", data=factors_df, palette="viridis", legend=False)
plt.title("Ringkasan Faktor yang Paling Berpengaruh terhadap Stroke")
plt.xlabel("Proporsi Stroke")
plt.ylabel("Faktor")
plt.show()
```
