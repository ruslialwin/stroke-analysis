```python
# persentase stroke distribution
stroke_counts = df['stroke'].value_counts()
plt.pie(stroke_counts, labels=stroke_counts.index,autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'] )
plt.title('Stroke Distribution')
```

```python
# jumlah data dari setiap variasi data pada kolom
fig,axes = plt.subplots(4,2,figsize = (16,16))
sns.set_style('whitegrid')
fig.suptitle("Count plot for various categorical features")

sns.countplot(ax=axes[0,0],data=df,x='gender')
sns.countplot(ax=axes[0,1],data=df,x='hypertension')
sns.countplot(ax=axes[1,0],data=df,x='heart_disease')
sns.countplot(ax=axes[1,1],data=df,x='ever_married')
sns.countplot(ax=axes[2,0],data=df,x='work_type')
sns.countplot(ax=axes[2,1],data=df,x='Residence_type')
sns.countplot(ax=axes[3,0],data=df,x='smoking_status')
sns.countplot(ax=axes[3,1],data=df,x='stroke')

plt.show()
```

```python
# melihat hubungan `work_type` dengan `avg_glucose_level`
df.groupby(['work_type'])['avg_glucose_level'].median().plot(kind = 'bar')
plt.show()
```

```python
# pesebaran data tipe pekerjaan
fig = plt.figure(figsize=(8, 6))
ax = sns.countplot(y=df['work_type'])

plt.title("work type")
plt.xlabel("Jumlah")
plt.ylabel("pekerjaan")
```

```python
# Bivariate
# Filter data to include only rows where 'stroke' is 1
df_stroke = df[df['stroke'] == 1]

# Create subplots for smoking_status, gender, hypertension, heart_disease, work_type, ever_married, and Residence_type
plt.figure(figsize=(30, 16))  # Adjusting the figure height

# Subplot 1: gender
plt.subplot(2, 4, 1)
sns.countplot(x='gender', hue='stroke', data=df, palette='viridis')
plt.title('gender vs stroke')
plt.xlabel('Gender')
plt.ylabel('Count')

# Subplot 2: Hypertension
plt.subplot(2, 4, 2)
sns.countplot(x='hypertension', hue='stroke', data=df, palette='viridis')
plt.title('hypertension vs stroke')
plt.xlabel('Hypertension')
plt.ylabel('Count')

# Subplot 3: Heart disease
plt.subplot(2, 4, 3)
sns.countplot(x='heart_disease', hue='stroke', data=df, palette='viridis')
plt.title('Heart Disease vs stroke')
plt.xlabel('Heart Disease')
plt.ylabel('Count')

# Subplot 4: ever married
plt.subplot(2, 4, 4)
sns.countplot(x='ever_married', hue='stroke', data=df, palette='viridis')
plt.title('Ever Married vs stroke')
plt.xlabel('Ever married')
plt.ylabel('Count')

# Subplot 5: Work Type
plt.subplot(2, 4, 5)
sns.countplot(x='work_type', hue='stroke', data=df, palette='viridis')
plt.title('Work Type vs stroke')
plt.xlabel('work type')
plt.ylabel('Count')

# Subplot 6: Residence type
plt.subplot(2, 4, 6)
sns.countplot(x='Residence_type', hue='stroke', data=df, palette='viridis')
plt.title('Residence type vs stroke')
plt.xlabel('Residence type')
plt.ylabel('Count')

# Subplot 7: smoking status
plt.subplot(2, 4, 7)
sns.countplot(x='smoking_status', hue='stroke', data=df, palette='viridis')
plt.title('Smoking status vs stroke')
plt.xlabel('Smoking status')
plt.ylabel('Count')

# Adjust layout
plt.tight_layout()

# Show the combined chart
plt.show()
```

```python
#univariate
# Filter data to include only rows where 'stroke' is 1
df_stroke = df[df['stroke'] == 1]

# Create subplots for smoking_status, gender, hypertension, heart_disease, work_type, ever_married, and Residence_type
plt.figure(figsize=(30, 16))  # Menyesuaikan tinggi figur

# Subplot 1: Smoking Status
plt.subplot(2, 4, 1)
sns.countplot(x='smoking_status', data=df_stroke, palette='viridis', hue='stroke')
plt.title('Stroke Count by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')

# Subplot 2: Gender
plt.subplot(2, 4, 2)
sns.countplot(x='gender', data=df_stroke, palette='viridis', hue='stroke')
plt.title('Stroke Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Subplot 3: Hypertension
plt.subplot(2, 4, 3)
sns.countplot(x='hypertension', data=df_stroke, palette='viridis', hue='stroke')
plt.title('Stroke Count by Hypertension')
plt.xlabel('Hypertension')
plt.ylabel('Count')

# Subplot 4: Heart Disease
plt.subplot(2, 4, 4)
sns.countplot(x='heart_disease', data=df_stroke, palette='viridis', hue='stroke')
plt.title('Stroke Count by Heart Disease')
plt.xlabel('Heart Disease')
plt.ylabel('Count')

# Subplot 5: Work Type
plt.subplot(2, 4, 5)
sns.countplot(x='work_type', data=df_stroke, palette='viridis', hue='stroke')
plt.title('Stroke Count by Work Type')
plt.xlabel('Work Type')
plt.ylabel('Count')

# Subplot 6: Ever Married
plt.subplot(2, 4, 6)
sns.countplot(x='ever_married', data=df_stroke, palette='viridis', hue='stroke')
plt.title('Stroke Count by Ever Married')
plt.xlabel('Ever Married')
plt.ylabel('Count')

# Subplot 7: Residence Type
plt.subplot(2, 4, 7)
sns.countplot(x='Residence_type', data=df_stroke, palette='viridis', hue='stroke')
plt.title('Stroke Count by Residence Type')
plt.xlabel('Residence Type')
plt.ylabel('Count')

# Adjust layout
plt.tight_layout()

# Show the combined chart
plt.show()
```
